## Código python para ver resultados en LSI

```python
import matplotlib.pyplot as plt
import numpy as np

def plot_suspensos_lsi():
    # Datos de ejemplo: Prácticas en LSI y cantidad de suspensos para 200 alumnos
    practicas = ['Práctica 1', 'Práctica 2', 'Práctica 3']
    suspensos_practica_1 = np.random.randint(20, 50, 200)  # Simulación de datos para la práctica 1
    suspensos_practica_2 = np.random.randint(20, 40, 200)   # Simulación de datos para la práctica 2
    suspensos_practica_3 = np.random.randint(10, 20, 200)    # Simulación de datos para la práctica 3

    # Asegurar que la práctica 3 tiene menos suspensos que la práctica 1
    suspensos_practica_3 = np.minimum(suspensos_practica_1, suspensos_practica_3)

    # Calcular el promedio de suspensos para cada práctica
    promedio_suspensos = [np.mean(suspensos_practica_1), np.mean(suspensos_practica_2), np.mean(suspensos_practica_3)]

    # Crear la gráfica de barras
    plt.bar(practicas, promedio_suspensos, color='red')
    plt.xlabel('Prácticas en LSI')
    plt.ylabel('Cantidad Promedio de Suspensos')
    plt.title('Suspensos en Prácticas de LSI para 200 Alumnos')
    plt.show()

if __name__ == "__main__":
    plot_suspensos_lsi()
```

## Campana de Gauss (por si hiciese falta)

```python
import numpy as np
import matplotlib.pyplot as plt

# Función que calcula la campana de Gauss
def gaussian(x, mu, sigma):
    return np.exp(-((x - mu) ** 2) / (2 * sigma ** 2)) / (sigma * np.sqrt(2 * np.pi))

# Generar datos
x = np.linspace(-5, 5, 1000)
y = gaussian(x, mu=0, sigma=1)

# Crear la gráfica
plt.plot(x, y, label='Campana de Gauss')
plt.title('Campana de Gauss')
plt.xlabel('X')
plt.ylabel('Densidad de probabilidad')
plt.legend()
plt.grid(True)
plt.show()
```

## Imágenes

```python
import numpy as np
import cv2
import matplotlib.pyplot as plt

# Crear una imagen sintética
width, height = 400, 300
synthetic_image = np.zeros((height, width, 3), dtype=np.uint8)

# Dibujar patrones en la imagen
synthetic_image[:, :width//2, 0] = 255  # Canal R: rectángulo azul en la mitad izquierda
synthetic_image[:, width//2:, 2] = 255  # Canal B: rectángulo rojo en la mitad derecha

# Convertir de RGB a YCrCb
synthetic_ycrcb = cv2.cvtColor(synthetic_image, cv2.COLOR_RGB2YCrCb)

# Obtener las componentes Y, Cr y Cb
synthetic_y = synthetic_ycrcb[:, :, 0]
synthetic_cr = synthetic_ycrcb[:, :, 1]
synthetic_cb = synthetic_ycrcb[:, :, 2]

# Visualizar las imágenes
plt.figure(figsize=(18, 10))  # Tamaño de la figura
plt.subplot(141), plt.imshow(synthetic_image), plt.title('RGB')  # Plot de la imagen RGB

# Plot de las componentes. Se utilizan mapas de color diferentes para cada una.
plt.subplot(142), plt.imshow(synthetic_y, cmap='gray', vmax=255, vmin=0), plt.title('Y (Luminancia)')
plt.subplot(143), plt.imshow(synthetic_cr, cmap='RdYlBu', vmin=0, vmax=255), plt.title('Cr (Crominancia roja)')
plt.subplot(144), plt.imshow(synthetic_cb, cmap='YlGnBu', vmin=0, vmax=255), plt.title('Cb (Crominancia azul)')
plt.show()
```

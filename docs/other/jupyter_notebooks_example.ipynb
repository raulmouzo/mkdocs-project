# Jupyter Notebooks

Se pueden añadir archivos de jupyter notebooks fácilmente en MkDocs, de hecho, estás leyendo uno. 

## Instalación

```
pip install mkdocs-jupyter

```

## Configuración

Para configurar este complemento debemos añadir lo siguiente a `mkdocs.yml`
```
nav:
    - Home: index.md
    - Notebook page: notebook.ipynb
    - Python file: python_script.py
plugins:
    - mkdocs-jupyter
```

Para más información consulta el repositorio del autor [aquí](https://github.com/danielfrg/mkdocs-jupyter).

## Ejemplo

````python
# Importar bibliotecas necesarias
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Configuración para mostrar gráficos en el notebook
%matplotlib inline

# Crear datos de ejemplo
x = np.linspace(0, 2*np.pi, 100)
y1 = np.sin(x)
y2 = np.cos(x)

# Crear un gráfico con dos líneas
plt.figure(figsize=(10, 6))
plt.plot(x, y1, label='Seno')
plt.plot(x, y2, label='Coseno')
plt.title('Gráfico de Seno y Coseno')
plt.xlabel('X')
plt.ylabel('Y')
plt.legend()
plt.grid(True)
plt.show()

# Crear una tabla de datos usando pandas
data = {'X': x, 'Seno': y1, 'Coseno': y2}
df = pd.DataFrame(data)
df.head(10)  # Mostrar las primeras 10 filas de la tabla

# Realizar operaciones matemáticas
media_seno = np.mean(y1)
suma_coseno = np.sum(y2)

# Mostrar resultados de operaciones
print(f'Media del Seno: {media_seno:.2f}')
print(f'Suma del Coseno: {suma_coseno:.2f}')
````
